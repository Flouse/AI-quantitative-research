```python
# for more information about finance.run_query(), please visit https://www.joinquant.com/help/api/help?name=JQData
def get_fund_stock(fund_stock_df):
    # 仅获取2020年第一季度基金持股信息
    fund_stock_list = []
    for i in range(10000):
        code=str(i)
        code = '0'*(6-len(code)) + code
        print('\r' + code, end='', flush=True)
        df = finance.run_query(query(finance.FUND_PORTFOLIO_STOCK).filter(finance.FUND_PORTFOLIO_STOCK.code==code,
                                                                          finance.FUND_PORTFOLIO_STOCK.report_type_id=='403001').limit(1000))

        df = df[df['proportion']!=0]
        if not df.empty:
            df.sort_values(by='proportion', ascending=False)
            df['year']=df['period_start'].map(lambda x:x.year)
            df = df[df['year']==2020]
            fund_stock_list.append(df)
    fund_stock_df = pd.concat(fund_stock_list,ignore_index=True)
    return fund_stock_df
```

```python
fund_stock_df=get_fund_stock(fund_stock_df)
```
