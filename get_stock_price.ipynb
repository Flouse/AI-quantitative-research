This file is used to get the stock price. For each stock we get the price between 2020/01/01 to 2020/07/31.

```python
from jqdatasdk import *
import numpy as np
import pandas as pd
```

```python
# please register on https://www.joinquant.com/help/api/help?name=JQData
auth('account','password')
```

```python
df = pd.read_csv('Data/fund_stock_2020_10000.csv',index_col=0)
```

```python
df
```

```python
symbol_set = set()
```

```python
# save stock symbol and name
for i in range(len(df)):
    symbol = df.loc[i,['symbol']][0]
    name = df.loc[i,['name']][0]
    symbol_set.add((symbol,name))
```

```python
len(symbol_set)
```

```python
normalized_set = set()
error_set = set()
```

```python
# get the normalized code by using function normalize_code()
# for more information about function normalize_code(), please visit https://www.joinquant.com/help/api/help?name=JQData
# for example, '601997' is the original stock code, and '601997.XSHG' is the normalized code
i = 0
for symbol, name in list(symbol_set):
    i += 1
    print(i)
    try:
        normalized_symbol = normalize_code(symbol)
    except:
        error_set.add((symbol,name))
    else:
        normalized_set.add((normalized_symbol,symbol,name))
```

```python
list(normalized_set)[0]
```

```python
# get the stock price for each stock in normalized_set
stock_price_list = []
i = 0
for symbol,_,name in list(normalized_set):
    i += 1
    print(i)
    stock_df = get_price(security = symbol, start_date="2020-01-01", end_date="2020-07-31", frequency='daily')
    stock_df['code'] = symbol
    stock_df['name'] = name
    stock_df['time'] = stock_df.index
    stock_price_list.append(stock_df)
stock_price_df = pd.concat(stock_price_list,ignore_index=True)
```

```python
stock_price_df
```

```python
# save dataframe
stock_price_df.to_csv('Data/stock_price_2020,csv')
```
