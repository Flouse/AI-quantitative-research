本文件用于获取3种基线指数：上证50、中证500、沪深300。

日期范围为2020-01-01至2020-07-31。


This file is used to get 3 baseline indexes: SSE 50, CSI 500, CSI 300. 

The date starts from 2020/01/01 and ends at 2020/07/31.

```python
from jqdatasdk import *
# please register on https://www.joinquant.com/help/api/help?name=JQData
auth('account','password')
```

```python
start_date = "2020-01-01"
end_date = "2020-07-31"
```

```python
# 关于get_price()参数使用方法，请参考 https://www.joinquant.com/help/api/help?name=JQData
# For more information about get_price(), please visit https://www.joinquant.com/help/api/help?name=JQData
shangzheng50_df = get_price(security = "000016.XSHG", start_date=start_date, end_date=end_date, frequency='daily')
shangzheng50_df['code'] = '000016.XSHG'
shangzheng50_df['name'] = '上证50'
shangzheng50_df
```

```python
zhongzheng500_df = get_price(security = "000905.XSHG", start_date=start_date, end_date=end_date, frequency='daily')
zhongzheng500_df['code'] = "000905.XSHG"
zhongzheng500_df['name'] = "中证500"
zhongzheng500_df
```

```python
hushen300_df = get_price(security = "000300.XSHG", start_date=start_date, end_date=end_date, frequency='daily')
hushen300_df['code'] = "000300.XSHG"
hushen300_df['name'] = "沪深300"
hushen300_df
```

```python
# save dataframe
shangzheng50_df.to_csv('shangzheng50_2020.csv')
zhongzheng500_df.to_csv('zhongzheng500_2020.csv')
hushen300_df.to_csv('hushen300_2020.csv')
```
